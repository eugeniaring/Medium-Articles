## 5. Differentiation in Autograd

```python
# Define operations
x1 = torch.tensor([2.0], requires_grad=True).float()
z1 = x1

x2 = torch.tensor([2.0], requires_grad=True).float()
z2 = x2**2

x3 = torch.tensor([2.0], requires_grad=True).float()
z3 = x3**3
```

```python
# Backward
z1.backward()
z2.backward()
z3.backward()
```

```python
from matplotlib.pyplot import *
from numpy import *

fig,ax=subplots(1,3,figsize=(15,4))
x=linspace(-1,1,5000)
y1=x
y2=x**2
y3=x**3

ax[0].plot(x,y1)
ax[1].plot(x,y2)
ax[2].plot(x,y3)
```

## 6. Introduction to Neural Networks

We are going to study fully connected neural network, the simplest model of neural networks. The goal of the hidden layers is to extract useful features, while the goal of the output layer is to predict the values of the target variable, that can be categorical or numerical. 
Each neuron in a layer is connected with every neuron in the next layer. The connections are called weights and are represented by a tensor matrix.  

### First example

```python
import torch

input_layer = torch.rand(10)
print(input_layer.shape)

#h1 contains 20 units
w1 = torch.rand(10,20)
b1 = torch.randn(20)
#h2 contains 20 units
w2 = torch.rand(20,20)
b2 = torch.randn(20)
#the output layer has 10 units
w3 = torch.rand(20,10)
b3 = torch.randn(10)

h1 = torch.matmul(w1.T,input_layer)+b1
print("First hidden layer:")
print(h1,"\n")
h2 = torch.matmul(w2.T,h1)+b2
print("Second hidden layer:")
print(h2,"\n") 
output_layer = torch.matmul(w3.T,h2)+b3
print("Output layer:")
print(output_layer)
```

```python
import torch 
import torch.nn as nn

#class Net inherits from nn.Module
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        # nn.Linear(i,o) where i=# units in the current layer, o = # units in the next layer
        self.fc1 = nn.Linear(10,20)
        self.fc2 = nn.Linear(20,20)
        self.out = nn.Linear(20,10)
        
    def forward(self,x):
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.out(x)
        return x
```

```python
input_layer = torch.rand(10)
#instantiate model
net = Net()
result = net(input_layer)
print(result)
```

## 7. Activation function

```python
def activation(x):
    return 1/(1+torch.exp(-x))

# Flatten the input images
images = torch.rand(1,28,28)
inputs = images.view(images.shape[0], -1)

# Create parameters
w1 = torch.randn(784, 256)
b1 = torch.randn(256)

w2 = torch.randn(256, 10)
b2 = torch.randn(10)

h = activation(torch.matmul(inputs, w1) + b1)

out = torch.matmul(h, w2) + b2
print(out)
```

```python
## Solution
def softmax(x):
    return torch.exp(x)/torch.sum(torch.exp(x), dim=1).view(-1, 1)

probabilities = softmax(out)

# Does it have the right shape? Should be (64, 10)
print(probabilities.shape)
# Does it sum to 1?
print(probabilities.sum(dim=1))
```

```python
from torch import nn

class Network(nn.Module):
    def __init__(self):
        super().__init__()
        
        # Inputs to hidden layer linear transformation
        self.hidden = nn.Linear(784, 256)
        # Output layer, 10 units - one for each digit
        self.output = nn.Linear(256, 10)
        
        # Define sigmoid activation and softmax output 
        self.sigmoid = nn.Sigmoid()
        self.softmax = nn.Softmax(dim=1)
        
    def forward(self, x):
        # Pass the input tensor through each of our operations
        x = self.hidden(x)
        x = self.sigmoid(x)
        x = self.output(x)
        x = self.softmax(x)
        
        return x
```
